```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import Normalize

Test_Data = pd.DataFrame({
    **{'L0_' +  str(i): np.full(10, 4) if i in [] else np.full(10, 1) if i in [5] else np.full(10, 15) for i in range(1, 10)},
    **{'L1_' + str(i): np.full(10, 2) if i in [18, 19, 17, 20] else (np.full(10, 5) if i in [15,16, 21, 22] else np.full(10, 15)) for i in range(1, 37)},
    **{'L2_' + str(i): np.full(10, 19) if i in [] else (np.full(10, 3) if i in [18,19,17, 20] else (np.full(10, 7) if i in [16, 21] else np.full(10, 17))) for i in range(1, 37)},
    **{'L3_' + str(i): np.full(10, 4) if i in [] else np.full(10, 8) if i in [5] else np.full(10, 15) for i in range(1, 10)},
    **{'Had_' + str(i): np.full(10, 9) if i in [2, 4, 6, 8] else np.full(10, 18) if i in [5] else np.full(10, 6) for i in range(1, 10)}
})


# Function to reshape columns into a matrix
def reshape_to_matrix(columns, shape):
    data = Test_Data[columns]
    matrices = data.values.reshape(-1, *shape)
    return list(matrices)

# Reshape the data into matrices and store them in the DataFrame
Test_Data['L0_matrix'] = reshape_to_matrix(['L0_' + str(i) for i in range(1, 10)], (3, 3))
Test_Data['L1_matrix'] = reshape_to_matrix(['L1_' + str(i) for i in range(1, 37)], (3, 12))
Test_Data['L2_matrix'] = reshape_to_matrix(['L2_' + str(i) for i in range(1, 37)], (3, 12))
Test_Data['L3_matrix'] = reshape_to_matrix(['L3_' + str(i) for i in range(1, 10)], (3, 3))
Test_Data['Had_matrix'] = reshape_to_matrix(['Had_' + str(i) for i in range(1, 10)], (3, 3))

# Function to plot a matrix with a square grid aspect ratio
def plot_square_grid_matrix(matrix, title, matrix_type):
    fig, ax = plt.subplots(figsize=(6, 6))  # Set figure size to be square

    # Determine the aspect ratio based on matrix type
    if matrix_type in ['L0', 'L3', 'Had']:
        aspect = 1.0  # Square cells for 3x3 matrices
    elif matrix_type in ['L1', 'L2']:
        aspect = 12 / 3  # Compress the width for L1 and L2

    # Plot the matrix
    norm = Normalize(vmin=0, vmax=19)
    cax = ax.matshow(matrix, cmap='tab20', norm=norm, aspect=aspect)
    plt.title(title)

    # Adding grid lines based on the matrix size
    nrows, ncols = matrix.shape
    for i in range(nrows-1):
        ax.axhline(i+0.5, color='white', linestyle='-', linewidth=2)
    for i in range(ncols-1):
        ax.axvline(i+0.5, color='white', linestyle='-', linewidth=2)

    # Turn off the axis labels
    ax.axis('off')

    # Adjust layout
    plt.tight_layout()
    plt.show()
# Plot the first matrix from each type with the new aspect ratio
plot_square_grid_matrix(Test_Data['L0_matrix'][0], 'L0 Matrix', 'L0')
plot_square_grid_matrix(Test_Data['L1_matrix'][0], 'L1 Matrix', 'L1')
plot_square_grid_matrix(Test_Data['L2_matrix'][0], 'L2 Matrix', 'L2')
plot_square_grid_matrix(Test_Data['L3_matrix'][0], 'L3 Matrix', 'L3')
plot_square_grid_matrix(Test_Data['Had_matrix'][0], 'Had Matrix', 'Had')
```
